```python
x = np.arange(len(csv1_copy))

values = csv1_copy['합계']

plt.figure(figsize = (20, 10))
plt.bar(x, values)

plt.show()
```

```python
#벼가 잘 자라기 위해서는 연간 강수량이 1000mm 이상이어야함
#서리 안 내리는 날 150일 이상
#벼 자라는 동안 온도 17~18도 유지
#9, 10월에 비가 많이 오면 안됨
#비가 많이 안 오는 기준을 평균값 미만으로 설정
```

```python
csv1_copy = csv1_copy[csv1_copy['합계'] >= 1000]
```

```python
csv1_copy = csv1_copy[csv1_copy['9월 강수량'] < csv1_copy['9월 강수량'].mean()]
```

```python
csv1_copy = csv1_copy[csv1_copy['10월 강수량'] < csv1_copy['10월 강수량'].mean()]
```

```python
csv1_copy = csv1_copy.sort_values('합계', ascending=True)
```

```python
x = np.arange(len(csv1_copy))

names = csv1_copy['관측소']
values = csv1_copy['합계']

plt.figure(figsize = (12, 8))
plt.barh(x, values)
plt.yticks(x, names)
plt.xlim(1000, 1400)
plt.xlabel('강수량(mm)')

plt.show()
```

```python
type(names)
```

###전라북도 14개 시군에서 특정 조건을 만족하는 시군을 추출하여 2019년 1년동안의 강수량을 분석한 그래프.

####특정 조건이란 벼가 잘 자라는 조건의 강수량을 보유하는 시군이며, 벼가 잘 자라는 조건은 연간 강수량 1000mm이상, 그 중 수확철인 9월과 10월은 강수량이 저조(평균값 미만일 경우 조건을 만족한다고 가정)할 때이다.

####두 조건을 만족하는 시군 중 진안 정천면과 장수 뜬봉샘, 임실 강진면까지가 강수량이 많은 지역임을 확인할 수 있다.
####그래프에 나온 지역 모두 두 조건을 만족하지만 강수량으로 봤을 때 그래프 상 좌측에 있는 지역일수록 벼가 잘 자랄 확률이 높은 지역이라고 예측할 수 있다.

```python
csv5 = pd.read_csv('제주특별자치도_연도별노인인구현황_20210630.csv', encoding='cp949')
```

```python
csv5.head()
```

```python
csv5_pop = csv5['도인구']
csv5_pop = csv5_pop.sort_values(ascending=True)
```

```python
csv5_65 = csv5['65세이상 제주시 인구'] + csv5['65세이상 서귀포시 인구']
csv5_65 = csv5_65.sort_values(ascending=True)
```

```python
csv5_80 = csv5['80세이상 제주시 인구'] + csv5['80세이상 서귀포시 인구']
csv5_80 = csv5_80.sort_values(ascending=True)
```

```python
fig, axes = plt.subplots(3, 1)
x = np.arange(2006, 2021)

axes[0].plot(x, csv5_pop)
axes[0].set_title('2006~2020년 제주도 인구 수')
axes[0].set_ylabel('인구 수(명)')

axes[1].plot(x, csv5_65, 'r')
axes[1].set_title('2006~2020년 제주도 65세 이상 인구 수')
axes[1].set_ylabel('인구 수(명)')

axes[2].plot(x, csv5_80, 'g')
axes[2].set_title('2006~2020년 제주도 80세 이상 인구 수')
axes[2].set_ylabel('인구 수(명)')


plt.tight_layout(pad=1.5)
plt.show()
```
